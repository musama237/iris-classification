**The Iris Flower Classification project is an excellent starting point for beginners in machine learning. It's based on the Iris flower dataset, one of the most famous datasets in the field**

```python
#loading libraries
import numpy
import pandas as pd
import matplotlib
import sklearn
```

```python
#loading the dataset
from sklearn import datasets
iris = datasets.load_iris()
iris_df = pd.DataFrame(data = iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df.head()
```

```python
print(iris_df.isnull().sum())
```

```python
#split the data into test_train
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state= 0)

#scaling the data
scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.fit_transform(X_test)
```

```python
#using Knn to classify
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=5)
model = model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix

# Confusion Matrix
print(confusion_matrix(y_test, y_pred))

# Classification Report
print(classification_report(y_test, y_pred))
```
